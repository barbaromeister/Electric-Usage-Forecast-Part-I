```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\ASUS\Models\tic_ag.csv')
```

```python
df
```

```python
table = df.pivot_table(
    index='PROFILEDATE',
    columns='POINTSERNO',
    values='IMPORTNETTING'
)
```

```python
table.index = pd.to_datetime(table.index.astype(str), format='%Y%m%d%H%M%S')  
```

```python
import random
plt.figure(figsize=(25, 5))

# 10 farklÄ± POINTSERNO seÃ§mek iÃ§in dÃ¶ngÃ¼
for i in range(10):
    number = random.choice(table.columns)  # Rastgele bir sÃ¼tun adÄ± seÃ§
    

    plt.figure(figsize=(25, 5))

    plt.plot(table.index, table[number], label=f'POINTSERNO {number}')  # SeÃ§ilen sÃ¼tunu Ã§iz

    plt.title('Usage of Electricity')
    plt.xlabel('Time')
    plt.ylabel('Electricity Usage')
    plt.legend()
    plt.xticks(rotation=90)  # X eksenindeki zaman etiketlerini dÃ¶ndÃ¼r
    plt.grid(True)
    plt.show()
```

```python
sum_table = table.sum(axis=1)
plt.figure(figsize=(25, 5))
plt.plot(sum_table.index, sum_table, label='Total Electricity Usage')
plt.title('Total Electricity Usage')
plt.xlabel('Time')
plt.ylabel('Electricity Usage')
plt.legend()
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
